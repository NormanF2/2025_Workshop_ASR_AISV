## Italian Automatic Speech Recognition with NVIDIA Conformer CTC

Pre-trained model from Nvidia available at this [link](https://huggingface.co/nvidia/stt_it_conformer_ctc_large)
*   Encoder: [Conformer](https://sh-tsang.medium.com/brief-review-conformer-convolution-augmented-transformer-for-speech-recognition-88dbf40240db)
*   Decoder: [CTC](https://medium.com/@akp83540/connectionist-temporal-classification-ctc-722bbb767e62#:~:text=Connectionist%20Temporal%20Classification%20is%20a%20powerful%20technique%20for%20training%20RNNs,process%20and%20achieving%20impressive%20accuracy.)

```python
# Environement preparation
wav_path = './wav_test.wav'
target = 'Si tratta principalmente di spiagge sabbiose, sicure per il nuoto, e la maggior parte di esse gode dell\'ombra degli alberi di pohutukawa.'
clean_target = 'si tratta principalmente di spiagge sabbiose sicure per il nuoto e la maggior parte di esse gode dell\'ombra degli alberi di pohutukawa'
```

```python
# Pre-trained Model instantiation through Nvidia Nemo library
import nemo.collections.asr as nemo_asr
asr_model = nemo_asr.models.EncDecCTCModelBPE.from_pretrained("nvidia/stt_it_conformer_ctc_large")
```

Upload data

```python
# Listen to your WAV file
import IPython
IPython.display.Audio(wav_path)
```

```python
# Transcription
hyp = asr_model.transcribe([wav_path])
```

```python
# Printing result against target
print(f'TARGET: {clean_target}')
print(f'Hyp:    {hyp[0]}')
```

```python
# Metrics measurements
import jiwer
wer = jiwer.wer(target, hyp[0])
cer = jiwer.cer(target, hyp[0])
print(f'WER: {wer}')
print(f'CER: {cer}')
```
